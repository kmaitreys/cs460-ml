```python
import pandas as pd

GDP = pd.read_csv('/home/ws10/Documents/GDP_interp.csv')

# print(GDP.loc[0][1:])

# Get number of rows in the dataframe
num_of_rows = len(GDP)

countries = []

for i in range(num_of_rows):
    # values = np.array(GDP.loc[i][1:])
    var_name = f'GDP_{GDP.loc[i][0]}'
    var_name = var_name.replace(' ', '_')
    # globals()[var_name] = values
    countries.append(var_name)


for i, country in enumerate(countries):
    globals()[country] = np.array(GDP.loc[i][1:])


x_arr = np.array(GDP.columns[1:], dtype=np.int32)
print(x_arr)

# Find UnivariateSpline for each country
for country in countries:
    f = InterpolatedUnivariateSpline(x_arr, globals()[country], k=2)
    globals()[f'{country}_interp'] = f

x_pred = np.linspace(2000, 2050, num=51)

# Predict GDP for each country
for country in countries:
    f = globals()[f'{country}_interp']
    globals()[f'{country}_pred'] = f(x_pred)


# Write the predicted to a new csv file
with open('GDP_pred.csv', 'w') as f:
    f.write('Country,2029,2030,2031,2032,2033,2034,2035,2036,2037,2038,2039,2040,2041,2042,2043,2044,2045,2046,2047,2048,2049,2050\n')
    for country in countries:
        f.write(f'{country[4:]},')
        for pred in globals()[f'{country}_pred']:
            f.write(f'{pred},')
        f.write('\n')

# Plot the predicted GDP in different plots for each country
import matplotlib.pyplot as plt

for country in countries:
    plt.plot(x_arr, globals()[country], 'o', x_pred, globals()[f'{country}_pred'], '-')
    plt.title(country[4:])
    plt.xlabel('Year')
    plt.ylabel('GDP (in billion USD))')
    plt.savefig(f'{country[4:]}_GDP.png')
    plt.show()

    
```

### MACHINE LEARNIG MODEL TO PREDICT MARINE ENERGY 

```python
import numpy as np
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, LSTM

# define input shape
timesteps = 21
features = 3

# create random training data
X_train = np.random.rand(1000, timesteps, features)
Y_train = np.random.rand(1000, 1)

# define model architecture
model = Sequential()
model.add(LSTM(128, input_shape=(timesteps, features)))
model.add(Dense(1))

# compile model
model.compile(loss='mse', optimizer='adam')

# train model
model.fit(X_train, Y_train, epochs=50, batch_size=32)

# generate predictions for next 30 time steps
X_test = np.random.rand(1, timesteps, features)
Y_pred = model.predict(X_test)
for i in range(30):
    X_test[0, :-1, :] = X_test[0, 1:, :]
    X_test[0, -1, :] = Y_pred
    Y_pred = model.predict(X_test)
    print(Y_pred)

# Plot the predictions

import matplotlib.pyplot as plt

plt.plot(Y_pred)
plt.show()


```
