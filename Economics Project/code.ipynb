```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split



final_data = pd.read_csv('/home/ws10/maitrey/cs460-ml/Economics Project/final_data.csv', skiprows=1)

# remove the first column
final_data = final_data.iloc[:, 1:]

# print(final_data)
brazil_rows = [0, 6, 12, 18]
brazil_data = final_data.iloc[brazil_rows]

china_rows = [1, 7, 13, 19]
china_data = final_data.iloc[china_rows]

india_rows = [2, 8, 14, 20]
india_data = final_data.iloc[india_rows]

japan_rows = [3, 9, 15, 21]
japan_data = final_data.iloc[japan_rows]

usa_rows = [4, 10, 16, 22]
usa_data = final_data.iloc[usa_rows]

# extract the data
brazil_data = brazil_data.to_numpy()
china_data = china_data.to_numpy()
india_data = india_data.to_numpy()
japan_data = japan_data.to_numpy()
usa_data = usa_data.to_numpy()

X_brazil = brazil_data[0:3,:21]
X_china = china_data[0:3,:21]
X_india = india_data[0:3,:21]
X_japan = japan_data[0:3,:21]
X_usa = usa_data[0:3,:21]


# Reshape X_data by transposing it
X_brazil = X_brazil.T
X_china = X_china.T
X_india = X_india.T
X_japan = X_japan.T
X_usa = X_usa.T

y_brazil = brazil_data[3,:21]
y_china = china_data[3,:21]
y_india = india_data[3,:21]
y_japan = japan_data[3,:21]
y_usa = usa_data[3,:21]



# train-test split for Brazil
X_train_brazil, X_test_brazil, y_train_brazil, y_test_brazil = train_test_split(X_brazil, y_brazil, test_size=0.2, random_state=42)

# train-test split for China
X_train_china, X_test_china, y_train_china, y_test_china = train_test_split(X_china, y_china, test_size=0.2, random_state=42)

# train-test split for India
X_train_india, X_test_india, y_train_india, y_test_india = train_test_split(X_india, y_india, test_size=0.2, random_state=42)

# train-test split for Japan
X_train_japan, X_test_japan, y_train_japan, y_test_japan = train_test_split(X_japan, y_japan, test_size=0.2, random_state=42)

# train-test split for USA
X_train_usa, X_test_usa, y_train_usa, y_test_usa = train_test_split(X_usa, y_usa, test_size=0.2, random_state=42)


# create the linear regression model
model_brazil = LinearRegression()
model_china = LinearRegression()
model_india = LinearRegression()
model_japan = LinearRegression()
model_usa = LinearRegression()

# train the model
model_brazil.fit(X_train_brazil, y_train_brazil)
model_china.fit(X_train_china, y_train_china)
model_india.fit(X_train_india, y_train_india)
model_japan.fit(X_train_japan, y_train_japan)
model_usa.fit(X_train_usa, y_train_usa)

# test the model
y_pred_brazil = model_brazil.predict(X_test_brazil)
y_pred_china = model_china.predict(X_test_china)
y_pred_india = model_india.predict(X_test_india)
y_pred_japan = model_japan.predict(X_test_japan)
y_pred_usa = model_usa.predict(X_test_usa)

mse_brazil = mean_squared_error(y_test_brazil, y_pred_brazil)
mse_china = mean_squared_error(y_test_china, y_pred_china)
mse_india = mean_squared_error(y_test_india, y_pred_india)
mse_japan = mean_squared_error(y_test_japan, y_pred_japan)
mse_usa = mean_squared_error(y_test_usa, y_pred_usa)

# Print mse as percentage of mean of y_test
print("Mean Squared Error for Brazil:", mse_brazil)
print("Mean Squared Error for China:", mse_china)
print("Mean Squared Error for India:", mse_india)
print("Mean Squared Error for Japan:", mse_japan)
print("Mean Squared Error for USA:", mse_usa)


# use the model for predictions
new_data_brazil = brazil_data[0:3,21:]
print(new_data_brazil.shape)
new_data_china = china_data[0:3,21:]
new_data_india = india_data[0:3,21:]
new_data_japan = japan_data[0:3,21:]
new_data_usa = usa_data[0:3,21:]

# Reshape new_data
new_data_brazil = new_data_brazil.T
new_data_china = new_data_china.T
new_data_india = new_data_india.T
new_data_japan = new_data_japan.T
new_data_usa = new_data_usa.T

predictions_brazil = model_brazil.predict(new_data_brazil)
predictions_china = model_china.predict(new_data_china)
predictions_india = model_india.predict(new_data_india)
predictions_japan = model_japan.predict(new_data_japan)
predictions_usa = model_usa.predict(new_data_usa)

# Concatenate predictions to original data
print(y_pred_brazil.shape, predictions_brazil.shape)
final_brazil = np.concatenate((brazil_data[3,:21], predictions_brazil))
final_china = np.concatenate((china_data[3,:21], predictions_china))
final_india = np.concatenate((india_data[3,:21], predictions_india))
final_japan = np.concatenate((japan_data[3,:21], predictions_japan))
final_usa = np.concatenate((usa_data[3,:21], predictions_usa))



# Plot the predictions and original data for each country
import matplotlib.pyplot as plt

plt.plot(final_brazil, label="Predictions")
plt.plot(brazil_data[3], label="Original Data")
plt.title("Brazil")
plt.legend()
plt.savefig('brazil_marine_pred.png')
plt.show()

plt.plot(final_china, label="Predictions")
plt.plot(china_data[3], label="Original Data")
plt.title("China")
plt.legend()
plt.savefig('china_marine_pred.png')
plt.show()

plt.plot(final_india, label="Predictions")
plt.plot(india_data[3], label="Original Data")
plt.title("India")
plt.legend()
plt.savefig('india_marine_pred.png')
plt.show()

plt.plot(final_japan, label="Predictions")
plt.plot(japan_data[3], label="Original Data")
plt.title("Japan")
plt.legend()
plt.savefig('japan_marine_pred.png')
plt.show()

plt.plot(final_usa, label="Predictions")
plt.plot(usa_data[3], label="Original Data")
plt.title("USA")
plt.legend()
plt.savefig('usa_marine_pred.png')
plt.show()
```
